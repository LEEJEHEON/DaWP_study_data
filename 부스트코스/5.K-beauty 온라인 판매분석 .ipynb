가설 : K-Beauty는 성장하고 있을까? 해외 직접판매를 한다면 어느 국가로 판매전략을 세우면 좋을까?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
sns.set(font="Malgun Gothic")
%config InlineBackend.figure_format = 'retina' #글씨 선명하게
```

```python
df_raw=pd.read_csv("국가_대륙_별_상품군별_온라인쇼핑_해외직접판매액_20200426015325.csv",encoding="cp949",engine='python')
```

```python
df_raw.shape
```

```python
df_raw["국가(대륙)별"].value_counts()
```

```python
df_raw[df_raw["국가(대륙)별"]=="미국"]
# boolean 데이터로 나오고 mask 해서 true만 데이터프레임으로 변경
# e = 추정치 , p : 잠정치, - : 자료없음, ...: 미상자료, x: 비밀보호
```

```python
# 컬럼을 행으로 바꿈
# 분기 년도별 데이터로 바꾸기 위해
# id_vars 만 정하면 나머지는 알아서 value_vars로 넘어감
df= df_raw.melt(id_vars=["국가(대륙)별","상품군별","판매유형별"], var_name="기간",value_name="백만원")
df.shape
# 행이 많이 늘어남
```

```python
df.head()
```

```python
df.info()
# 전처리를 할려면 숫자로 변형을 해줘야함
```

# 5.4 

```python
df["기간"]
```

```python
int("2019 4/4 p)".split()[0]) 
# p = 추정치
```

```python
df["연도"]=df["기간"].map(lambda x: int(x.split()[0]))
# map 은 안에 함수를 넣어줄 수 있음
# x를 받아서 x를 split 하고 0 인덱스만 출력
df.head() 
# map를 활용하여 익명함수를 만듬
```

```python
"2019 1/4 p)".split()[1].split("/")[0]
```

```python
df["분기"]=df["기간"].map(lambda x: int(x.split()[1].split("/")[0]))
df.head()
```

```python
df["백만원"]= df["백만원"].replace("-",pd.np.nan).astype(float)
# - 는 결측치를 의미 
# numpy를 import 안해도 pandas를 이용하여 접근 가능
# nan 값으로 변경 (타입을 변경하기 위해)
df.head()
```

```python
df.info()
```

```python
df=df[(df["국가(대륙)별"] != "합계") & (df["상품군별"] != "합계")].copy()
df.info()
# 필요없는 데이터 제거 
# 국가 대륙별 : 합계 = 모두 다 데이터가 합계임
# 상품군별 합계 :  2016년 이전은 데이터가 없기 때문에
# 메모리 용량이 줄어든 걸 확인할 수 있다.
```

```python
df.isnull().sum()
# 금액이 없는 데이터도 있다는 것을 확인
```

# 5.5

```python
df_total=df[df["판매유형별"]=="계"].copy()
df_total.head()
```

```python
sns.lineplot(data=df_total, x="연도",y="백만원")
#판매액이 증가 하고 있다는 것을 확인가능
```

```python
sns.lineplot(data=df_total, x="연도",y="백만원", hue="상품군별")

plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad=0.) #legend 값을 밖에 표시
#특히 화장품이 굉장히 많이 증가했다는 것을 확인
```

```python
sns.relplot(data=df_total, x="연도", y="백만원",hue="상품군별",kind="line",col="상품군별",col_wrap=4)
# 위에 것이랑 똑같이 나옴
# 화장품이 커서 다른 값들을 보기 어려움
```

```python
df_sub= df_total[~df_total["상품군별"].isin(["화장품"])].copy()
# 화장품만 뺴고 보기 위해
```

```python
sns.relplot(data=df_sub, x="연도", y="백만원",hue="상품군별",kind="line",col="상품군별",col_wrap=4)
# 화장품이 없이 보았을 때 의류, 패션이 굉장히 증가하고 있다는 것을 확인

```

```python
df_sub_f= df_total[~df_total["상품군별"].isin(["화장품","의류 및 패션관련 상품"])].copy()
```

```python
sns.relplot(data=df_sub_f, x="연도", y="백만원",hue="상품군별",kind="line",col="상품군별",col_wrap=4)
#가전 전자 통신기기도 증가
#음반 비디오 기기도 증가
#
```

# 5.6

```python
df_cosmetic=df_total[df_total["상품군별"] == "화장품" ].copy()
df_cosmetic["상품군별"].unique()
#화장품만 있다는 것을 확인
```

```python
plt.figure(figsize=(15,4))
sns.lineplot(data=df_cosmetic, x="연도",y="백만원",hue="분기")
# 4분기로 갈수록 증가 하고 있다.
# 이유는 계속 성장하고 있어서 그럼
```

```python
plt.figure(figsize=(15,4))
plt.xticks(rotation=30) # 글씨를 30도 정도 기울게 만든다.
sns.lineplot(data=df_cosmetic, x="기간",y="백만원")
# 꾸준히 증가해 왔다는 것을 알 수 있다.
```

```python
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_cosmetic, x="기간",y="백만원",hue="국가(대륙)별")
# 중국이 제일 판매량이 높다는 것을 알 수 있음
```

```python
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_cosmetic[df_cosmetic["국가(대륙)별"] != "중국"], x="기간",y="백만원",hue="국가(대륙)별")
# 아세한 , 미국 , 일본 순으로 나옴
```

```python
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
df_sub = df[df["판매유형별"] != "계"].copy()
sns.lineplot(data=df_sub,x="기간",y="백만원", hue="판매유형별")
# 2016년 이전에는 데이터가 없음
# 온라인 면세점이 훨씬 성장을 하고 있다
```

```python
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
df_sub = df[(df["판매유형별"] != "계") & (df["판매유형별"] != "면세점")].copy()
sns.lineplot(data=df_sub,x="기간",y="백만원", hue="판매유형별", ci=None)
# 전체적으로는 성장하고 있다고 보여짐
```

# 5.7

```python
df_fashion=df[(df["상품군별"] == "의류 및 패션관련 상품")&(df["판매유형별"]=="계")].copy()
df_fashion.head()
```

```python
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion, x="기간",y="백만원",hue="국가(대륙)별")
# 의류 및 패션관련 상품 쪽도 중국이 가장 높고
# 전체적으로는 성장 중이다.
```

```python
df_fashion_2=df[(df["상품군별"] == "의류 및 패션관련 상품")&(df["판매유형별"]!="계")].copy()
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion_2, x="기간",y="백만원",hue="판매유형별",ci=None)
# 2016년 부터는 데이터가 없어서 못그림
# 면세점 판매액이 주춤했다가 다시 성장을 했다.
```

```python
result=df_fashion.pivot_table(index="국가(대륙)별", columns="연도",values="백만원",aggfunc="sum")
#data를 앞에서 지정해도 상관없음
#기본값이 평균임
# 연도별 합계 금액을 표 형태로 구함
result
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(result, cmap="Blues_r",annot=True,fmt=".0f")
# 밝을 수록 많다.
# _r 을 하면 색 반전을 가능 (어두울수록 많게)
# 소수점없이 표현
```
