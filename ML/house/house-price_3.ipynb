```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import tensorflow as tf
%matplotlib inline
import seaborn as sns
```

```python
df_train=pd.read_csv("train.csv")
df_test=pd.read_csv("test.csv")
```

-----

```python
cal = ['MSZoning'
,'Alley'
,'LotShape'
,'LotConfig'
,'BldgType'
,'HouseStyle'
,'RoofStyle'
,'MasVnrType'
,'ExterQual'
,'Foundation'
,'BsmtQual'
,'BsmtCond'
,'BsmtExposure'
,'BsmtFinType1'
,'HeatingQC'
,'KitchenQual'
,'FireplaceQu'
,'GarageType'
,'GarageFinish'
,'GarageQual'
,'PoolQC'
,'Fence']
```

```python
x_train=df_train.loc[:,cal]
x_test=df_test.loc[:,cal]
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
add_x = pd.concat([x_train,x_test])
```

```python
one_hot = pd.get_dummies(add_x)
```

```python
train_onehot=one_hot.iloc[0:1460,:]
```

```python
test_onehot=one_hot.iloc[1460:2920,:]
```

----

```python
x_n =df_train.loc[:,["OverallQual","YearBuilt","YearRemodAdd","MasVnrArea","TotalBsmtSF","1stFlrSF","GrLivArea","FullBath","TotRmsAbvGrd","Fireplaces","GarageYrBlt","GarageCars","GarageArea"]]
```

```python
x_data_train = pd.concat([train_onehot,x_n], axis=1)
```

```python
x_data_train.shape
```

```python
x_data_train.isnull().sum().sort_values(ascending=False)
```

```python
x_data=x_data_train.drop("GarageYrBlt",axis=1)

x_data.loc[x_data.isnull()["MasVnrArea"],:] = 
        x_data.loc[x_data.isnull()["MasVnrArea"],:].fillna(
            value=x_data["MasVnrArea"].mean())
```

```python
x_data = x_data.values
```

```python
len(x_data)
```

```python
y_data = df_train.iloc[:,-1]
y_data.head()
```

```python
y_data = y_data.values
y_data=y_data.reshape(-1,1)
```

```python
print(x_data.shape)
print(y_data.shape)
```

```python
X = tf.placeholder(tf.float32, shape=[None,114])
Y = tf.placeholder(tf.float32, shape=[None,1])


W = tf.Variable(tf.random_normal([114,1],name='weight'))
b = tf.Variable(tf.random_normal([1],name='bias'))

hypothesis = tf.matmul(X, W) + b

cost = tf.reduce_mean(tf.square(hypothesis-Y))

optimizer = tf.train.GradientDescentOptimizer(learning_rate = 1e-8)
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
for step in range(200001):
    cost_val, hy_val, _ = sess.run(
        [cost,hypothesis,train],
        feed_dict={X:x_data, Y:y_data})
    if step % 2000 == 0:
        print(step,"Cost : ",cost_val,
             "\nPreadiction\n",hy_val)
```

```python
x_data.loc[x_data.isnull()["GarageCars"],:] = x_data.loc[x_data.isnull()["GarageCars"],:].fillna(value=x_data["GarageCars"].mean())
x_data.loc[x_data.isnull()["GarageArea"],:] = x_data.loc[x_data.isnull()["GarageArea"],:].fillna(value=x_data["GarageArea"].mean())
```

```python
x_data.loc[x_data.isnull()["TotalBsmtSF"],:] = x_data.loc[x_data.isnull()["TotalBsmtSF"],:].fillna(value=x_data["TotalBsmtSF"].mean())
```
