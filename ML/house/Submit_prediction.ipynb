```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
def log_rmse(predict_y, y_val):
    if predict_y.shape[0] != y_val.shape[0]:
        print('not matched length of arrays')
        return
    else:
        RMSE_log = np.sqrt(((np.log(predict_y)-np.log(y_val))**2).mean())
        return RMSE_log
```

```python
train_dataset = pd.read_csv("./house_prices/train.csv")
test_dataset = pd.read_csv("./house_prices/test.csv")
```

# 수치형 변수만 활용해 model 구축하기

```python
#Numeric variables
column_names = ['LotFrontage', 'LotArea', 'OverallQual', 'OverallCond',
       'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF',
       'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF', 'GrLivArea',
       'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath',
       'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces',
       'GarageYrBlt', 'GarageCars', 'GarageArea', 'WoodDeckSF',
       'OpenPorchSF', 'EnclosedPorch', '3SsnPorch', 'ScreenPorch',
       'PoolArea']
```

```python
continous_X_train = train_dataset.loc[:,column_names]
continous_X_train.head()
```

```python
continous_X_test = test_dataset.loc[:,column_names]
continous_X_test.head()
```

```python
# null값 확인 후 평균값으로 대체 진행
continous_X_train.isnull().sum().sort_values(ascending = False)
```

```python
continous_X_train['LotFrontage'] = continous_X_train['LotFrontage'].fillna(continous_X_train['LotFrontage'].mean())
continous_X_train['GarageYrBlt'] = continous_X_train['GarageYrBlt'].fillna(continous_X_train['GarageYrBlt'].mean())
continous_X_train['MasVnrArea'] = continous_X_train['MasVnrArea'].fillna(continous_X_train['MasVnrArea'].mean())
```

```python
# null값 확인 후 평균값으로 대체 진행
continous_X_test.isnull().sum().sort_values(ascending = False)
```

```python
continous_X_test['LotFrontage'] = continous_X_test['LotFrontage'].fillna(continous_X_test['LotFrontage'].mean())
continous_X_test['GarageYrBlt'] = continous_X_test['GarageYrBlt'].fillna(continous_X_test['GarageYrBlt'].mean())
continous_X_test['MasVnrArea'] = continous_X_test['MasVnrArea'].fillna(continous_X_test['MasVnrArea'].mean())
continous_X_test['BsmtFullBath'] = continous_X_test['BsmtFullBath'].fillna(continous_X_test['BsmtFullBath'].mean())
continous_X_test['BsmtHalfBath'] = continous_X_test['BsmtHalfBath'].fillna(continous_X_test['BsmtHalfBath'].mean())
continous_X_test['GarageCars'] = continous_X_test['GarageCars'].fillna(continous_X_test['GarageCars'].mean())
continous_X_test['BsmtFinSF1'] = continous_X_test['BsmtFinSF1'].fillna(continous_X_test['BsmtFinSF1'].mean())
continous_X_test['BsmtFinSF2'] = continous_X_test['BsmtFinSF2'].fillna(continous_X_test['BsmtFinSF2'].mean())
continous_X_test['BsmtUnfSF'] = continous_X_test['BsmtUnfSF'].fillna(continous_X_test['BsmtUnfSF'].mean())
continous_X_test['TotalBsmtSF'] = continous_X_test['TotalBsmtSF'].fillna(continous_X_test['TotalBsmtSF'].mean())
continous_X_test['GarageArea'] = continous_X_test['GarageArea'].fillna(continous_X_test['GarageArea'].mean())
```

```python
#Min-max normalization
for col in continous_X_train.columns:
    continous_X_train.loc[:,col] = (continous_X_train.loc[:,col] - continous_X_train.loc[:,col].min()) / (continous_X_train.loc[:,col].max() - continous_X_train.loc[:,col].min())
    
for col in continous_X_test.columns:
    continous_X_test.loc[:,col] = (continous_X_test.loc[:,col] - continous_X_test.loc[:,col].min()) / (continous_X_test.loc[:,col].max() - continous_X_test.loc[:,col].min()) 
```

```python
continous_X_test.isnull().sum()
```

```python
#reshape
y_train = train_dataset['SalePrice'].values
x_data = continous_X_train.values
y_data = np.array(y_train).reshape(-1,1)
print(x_data.shape, y_data.shape)
```

```python
# placeholders for a tensor that will be always fed.
X = tf.placeholder(tf.float32, shape=[None, x_data.shape[1]])
Y = tf.placeholder(tf.float32, shape=[None, 1])

W = tf.Variable(tf.random_normal([x_data.shape[1], 1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

# Hypothesis
hypothesis = tf.matmul(X, W) + b

# Simplified cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - Y))

# Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-4) #Tuning learning rate for gradient descent
train = optimizer.minimize(cost)

# Launch the graph in a session.
sess = tf.Session()
# Initializes global variables in the graph.
sess.run(tf.global_variables_initializer())

learning_cost = []

for step in range(100000): # Tuning # of Epochs
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train], 
                                   feed_dict={X: x_data, Y: y_data})
    learning_cost.append(cost_val)
    if step % 10000 == 0:
        print(step, "Cost:", cost_val, "\nPrediction:\n", hy_val)
```

```python
# 학습 성과 확인
log_rmse(hy_val.reshape(-1,), y_train)
```

```python
# test set 예측
predict_test = sess.run(hypothesis, feed_dict={X: continous_X_test})
```
