## Multivariable linear regression

Hypothesis <Br>
Cost function<br>
Gradient descent algorithm

여러개의 input이 있으면 Multivariable linear

1개의 input <br>
H(x) = Wx + b<br>
여러개(n개)의 input<br>
H(x1,x2,x3) = w1x1 + w2x2 + w3x3+...wnxn + b

input 이 많아 질수록 복잡해진다. <br>
그래서 Matrix를 이용한다. (곱셈을 이용) 
 > H(X)  = XW 

데이터를 인스턴스 라고 함

크기는 [x1 ,y1] , [x2,y2] 라고하면 <br>
 y1 과 x2 의 크기를 같아야 하고  곱셈을 하였을 때 결과는 <br>
 [x1 , y2] 의 크기로 나옴 <br>
 x1 = 데이터의 갯수<br>
 y1 = x 의 갯수 <br> 
 y2 = Y값 1 개 

이걸로 W = [x2,y2] 의 크기를 알아낼 수 있다.

실제 구현은 매트릭스를 많이 사용한다. <br>
매트릭스 사용하는것은H(x) = XW 이런 식으로 표현 

```python
import tensorflow as tf
```

```python
#데이터정의
x1_data = [73.,93.,89.,96.,73.]
x2_data = [80.,88.,91.,98.,66.]
x3_data = [75.,93.,90.,100.,70.]
y_data = [152.,185.,180.,196.,142.]
# fid_tict 로 하기 위한 선언
x1 = tf.placeholder(tf.float32)
x2 = tf.placeholder(tf.float32)
x3 = tf.placeholder(tf.float32)

Y = tf.placeholder(tf.float32)

w1 = tf.Variable(tf.random_normal([1]),name = 'weight1')
w2 = tf.Variable(tf.random_normal([1]),name = 'weight2')
w3 = tf.Variable(tf.random_normal([1]),name = 'weight3')
b = tf.Variable(tf.random_normal([1]),name = 'bias')

hypothesis = x1 * w1 + x2* w2 + x3*w3 +b 
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
#optimizer 을 0.01 or 0.001 을 하면 복잡해짐
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for step in range(2001):
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train],
                            feed_dict={x1:x1_data,x2:x2_data,x3:x3_data,Y: y_data})
    if step % 10 ==0:
        print(step, "Cost ",cost_val,"\nPrediction:\n",hy_val)
# 처음에는 cost가 높은데 실행할수록 작아지는 것을 확인
# 2000 Cost  3.1211815 
# Prediction:
# [154.32784 182.87228 181.94827 195.14641 140.93439]
# 마지막 값이 y값이랑 비숫해 진것을 확인 
# 지금은 3개라서 이정도 이지만 많아지면 복잡해지기에 거의 사용을 안함
```

```python
# Matrix 사용
# 각각이 x1, x2 ,x3 의 값
x_data = [[73.,80.,75.],[93.,88.,93.],
         [89.,91.,90.],[96.,98.,100.],[73.,66.,70.]]
y_data = [[152.],[185.],[180.],[196.],[142.]]

# shape x데이터는 n개를 가질 것이다 (지금은 5개임)
X = tf.placeholder(tf.float32, shape=[None, 3])
Y = tf.placeholder(tf.float32, shape=[None, 1])

w = tf.Variable(tf.random_normal([3,1]),name='weight')
b = tf.Variable(tf.random_normal([1]),name='bias')

hypothesis = tf.matmul(X, w)+ b

# 여기 부분만 간단해짐 ( Matrix 의 장점 )
```

```python
cost = tf.reduce_mean(tf.square(hypothesis- Y))

optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train= optimizer.minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for step in range(2001):
    cost_val, hy_val, _ = sess.run(
        [cost, hypothesis,train], feed_dict = {X : x_data, Y: y_data})
    if step % 10 == 0 :
        print(step, "Cost: ",cost_val,"\nPrediction\n",hy_val)
# 결과 값은 비슷해지는 것을 확인 
```
