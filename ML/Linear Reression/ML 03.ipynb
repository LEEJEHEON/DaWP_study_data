```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
x =[1,2,3]
y =[1,2,3] # 단순한 형태의 데이터가 주어짐
w= tf.placeholder(tf.float32) #여기있는 값을 바꾸기위해 
hypothesis = x*w #H(x)= Wx
```

```python
cost = tf.reduce_mean(tf.square(hypothesis-y))
sess = tf.Session()
sess.run(tf.global_variables_initializer())#초기화
w_val = []
cost_val = [] #그래프를 그리기 위해
```

```python
for i in range(-30,50): # -3 ~ 5 을 0.1씩 이동하겠다
    feed_w = i*0.1 
    curr_cost, curr_w = sess.run([cost,w],feed_dict={w:feed_w})
    w_val.append(curr_w) #0.1 씩 이동하면서 리스트에 넣음
    cost_val.append(curr_cost)
```

```python
plt.plot(w_val,cost_val) # x축은 w  , y축은 cost
plt.show()
```

```python
x_data =[1,2,3]
y_data =[1,2,3]
w= tf.Variable(tf.random_normal([1]),name='weight')
x = tf.placeholder(tf.float32)
y = tf.placeholder(tf.float32)
hypothesis = x*w
cost = tf.reduce_mean(tf.square(hypothesis-y))
```

```python
learning_rate= 0.1
gradient = tf.reduce_mean((w*x-y)*x)
descent = w-learning_rate*gradient
update = w.assign(descent)
# 복잡해지면 optimizer 로 해서 tensorflow에서 해줌
# optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.1)
# train= optimizer.minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for step in range(21):
    sess.run(update,feed_dict={x: x_data,y:y_data})
    print(step,sess.run(cost,feed_dict={x:x_data,y:y_data}),sess.run(w))
    # x 데이터와 y데이터를 주면서 update를 실행
    # 잘 되고 있는지 확인  x = 1 , y =1
    # step, cost, w 
    # cost는 계속 작아지고 w는 1로 가까워짐
```

```python
x =[1,2,3]
y =[1,2,3]
w= tf.Variable(5.0) # 말도 안되는 값을 주어짐 
# 잘 내려가는지 실행하기 위해
hypothesis = x*w
cost = tf.reduce_mean(tf.square(hypothesis-y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.1)
train= optimizer.minimize(cost)
# 계산식대신 이 함수를 사용
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for step in range(100):
    print(step,sess.run(w))
    sess.run(train)
```

gradient을 임의로 조정을 하고 싶다.

```python
x =[1,2,3]
y =[1,2,3]
w= tf.Variable(5.)  
hypothesis = x*w
gradient = tf.reduce_mean((w*x-y)*x)*2

cost = tf.reduce_mean(tf.square(hypothesis-y))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
# 여기까지는 똑같음
```

```python
# optimizer 에서 바로 minimizer를 안함
# cost에 맞는 greadient를 계산해 달라고 함
gvs= optimizer.compute_gradients(cost)
# 원하는대로 수정함
# 여기서는 바로 가서 minimizer랑 똑같음
apply_gradients = optimizer.apply_gradients(gvs)

sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
# 거의 같은 값이 나온다.
for step in range(100):
    print(step,sess.run([gradient,w,gvs]))
    sess.run(apply_gradients)
```
